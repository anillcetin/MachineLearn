```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,100,5)
y = np.sin(x)

plt.plot(x,y)

plt.title("Sine grafiği") #grafiğin başlığını yazıyoruz
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")

plt.show() #grafiği bastırıyoruz
```

```python
a = np.random.rand(50) #50 adet random sayı atıyor
b = np.random.rand(50)

colors = np.random.rand(50)

sizes = np.random.randint(50,150,50)

plt.scatter(a,b,c=colors,s=sizes)

plt.title("Nokta grafiği") #grafiğin başlığını yazıyoruz
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")

plt.show()
```

```python
data = np.random.randn(4000)

plt.hist(data, bins=30) #histogram grafiği şeklinde yapıyoruz.

plt.title("Histogram") #grafiğin başlığını yazıyoruz
plt.xlabel("Değerler")
plt.ylabel("Frekans")

plt.show()
```

```python
x = ['A','N','I','L']
y = np.random.randint(1,10,4) 

plt.bar(x,y)

plt.title("Sütun Grafiği") #grafiğin başlığını yazıyoruz
plt.xlabel("Harfler")
plt.ylabel("Değerler")

plt.show()
```

```python
c = [45,20,15,15,10,5]

plt.pie(c)

plt.title("Pasta dilimi grafiği")

plt.show()
```
